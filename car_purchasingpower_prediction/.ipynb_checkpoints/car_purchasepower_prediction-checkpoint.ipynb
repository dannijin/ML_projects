### Introduction
As a salesperson in a car dealership, the goal is to predict how much purchasing power a new customer would have (what car price the customer is likely to pay). It is a regression problem.

### load data

```python
import pandas as pd
import numpy as np
```

```python
data=pd.read_csv('Car_Purchasing_Data.csv', engine='python') # or use encoding = "ISO-8859-1"
```

```python
data.head(10)
```

```python
data.shape
```

### visualize data/data correlation

```python
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
sns.pairplot(data)
```

above shows car purchase amount is obviously positive correlated to age annual salary. Slightly correlated to net worth.

### clean up data, define input and target columns, normalize data, 

```python
# remove columns for customer name, email, country for now, and leave numerical data only
# remove target column

data_clean=data.drop(['Customer Name','Customer e-mail','Country','Car Purchase Amount'],axis=1)
```

```python
data_clean.head(10)
```

```python
#define input data X to train, target column y to predict

X=data_clean
y=data['Car Purchase Amount']
```

```python
X.shape
```

```python
y.shape
```

```python
#normalize X so that all attributes will have same scale

from sklearn.preprocessing import MinMaxScaler

scaler=MinMaxScaler()
X_scaled=scaler.fit_transform(X)
```

```python
X_scaled
```

```python
#normalize y
#need to reshape y so that y will have at least one column

y=y.values.reshape(-1,1)   # -1 as row means keep whatever number of rows in the original data
y_scaled=scaler.fit_transform(y)
```

```python
y.shape
```

```python
y_scaled
```

### train-test data split, define model,  fit train data

```python
from sklearn.model_selection import train_test_split

X_train,X_test, y_train, y_test= train_test_split(X_scaled, y_scaled, test_size=0.25)
```

```python
import tensorflow as tf
from tensorflow import keras

model = keras.Sequential ([
    keras.layers.Dense(units=25, input_shape=(5,), activation='relu'),
    keras.layers.Dense(units=25, activation='relu'),
    keras.layers.Dense(units=1, activation='linear')
])
```

```python
model.summary()
```

```python
model.compile(optimizer='adam', loss='mean_squared_error')
```

```python
history= model.fit(X_train, y_train, epochs=20, batch_size=20, verbose=1, validation_split=0.1)
```

### evaluate model

```python
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])

plt.xlabel('#epochs')
plt.ylabel('loss')
plt.title('compare training vs. validation loss over epochs')
plt.legend(['training loss','validation_loss'])
```

### predict on test data

```python
y_predict=model.predict(X_test)
```

```python
results = model.evaluate(X_test, y_test, batch_size=20)
print("test loss:", results)
```

```python
X_Testing = np.array([[0, 30, 300000, 0, 500000]])
y_predict_1=model.predict(X_Testing)
```
