### Introduction
Use Facebook Prophet to predict time series data: average price of avocado.

### Data cleaning

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from fbprophet import Prophet

%matplotlib inline

```

```python
data_raw=pd.read_csv('avocado.csv')
```

```python
data_raw.head(10)
```

```python
data_raw.isnull().sum()
```

```python
data=data_raw.sort_values('Date')
```

```python
data.head()
```

### Visualize data

```python
plt.figure(figsize=(20,10))
plt.plot(data['Date'], data['AveragePrice'])
plt.xlabel('Date')
plt.ylabel('Avocaco avg price')
```

```python
# check if data is balaced across different regions

data.groupby(['region']).count()
```

```python
plt.figure(figsize=(20,10))
sns.countplot(x='region', data=data)
plt.xticks(rotation=90);
```

### Train and predict

```python
# need to rename data as ds, target as y

data_1=data.rename(columns={'Date':'ds','AveragePrice':'y'})
```

```python
data_1=data_1[['ds','y']]
```

```python
data_1.head()
```

```python
model=Prophet()
model.fit(data_1)
```

```python
future=model.make_future_dataframe(periods=365)
prediction=model.predict(future)
```

```python
model.plot(prediction, xlabel='Date', ylabel='Avg price')
```

```python
model.plot_components(prediction)
```

overall predicted price is trending downward, with seasonality effect shows peak price around October.

### Select a particular region to predict

```python
# select region = Chicago to do the prediction
```

```python
data_chicago=data[data['region']=='Chicago']
```

```python
plt.figure(figsize=(20,10))
plt.plot(data_chicago['Date'], data_chicago['AveragePrice'])
```

```python
data_2=data_chicago[['Date','AveragePrice']]
```

```python
data_2=data_2.rename(columns={'Date':'ds','AveragePrice':'y'})
```

```python
model2=Prophet()
model2.fit(data_2)
```

```python
future2=model2.make_future_dataframe(periods=365)
pred_chicago=model2.predict(future2)
```

```python
model2.plot(pred_chicago)
```

```python
model2.plot_components(pred_chicago)
```

avocado price in Chicago has overall rising trend, with seasonality shows peak around end of October, and dip around late December.
